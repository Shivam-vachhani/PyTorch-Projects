```python
import torch
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df = df.drop(columns=['id','Unnamed: 32'])
df.head()
```

```python
df.shape
```

# Train Test Split

```python
X =df.drop(columns=['diagnosis'])
y = df['diagnosis']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

# Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

#Label Encoding

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

# Numpy Arrays to Tensors

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

# Making Nural Network

```python
class NuralNetwork():

  def __init__(self,X,layer_size=[4,4,4,1]):

    input_features = X.shape[1]
    self.layer_sizes = [input_features] + layer_size

    self.weights = []
    self.bias = []

    for i in range(len(self.layer_sizes) - 1):
      self.weights.append(torch.rand(self.layer_sizes[i+1], self.layer_sizes[i], dtype=torch.float64, requires_grad=True))
      self.bias.append(torch.zeros(self.layer_sizes[i+1], dtype=torch.float64, requires_grad=True))

  def forward(self, X):
    current_input = X

    for i in range(len(self.weights)):
      output_values = torch.matmul(current_input,self.weights[i].T)+ self.bias[i]
      current_input = torch.sigmoid(output_values)
    return current_input


  def loss_function(self,y_pred,y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)
    loss = -(y * torch.log(y_pred) + (1-y)*torch.log(1-y_pred)).mean()
    print(loss)
    return loss


```

```python

model = NuralNetwork(X_train_tensor)
epochs = 50
for epoch in range(epochs):
  y_pred = model.forward(X_train_tensor)
  loss = model.loss_function(y_pred,y_train_tensor)
  loss.backward()
  with torch.no_grad():
    for i in range(len(model.weights)):
      model.weights[i] -= 0.1 * model.weights[i].grad
      model.bias[i] -= 0.1 * model.bias[i].grad

      model.weights[i].grad.zero_()
      model.bias[i].grad.zero_()

print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
